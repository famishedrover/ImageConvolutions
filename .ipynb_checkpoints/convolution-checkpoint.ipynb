```python
def hadamard_prod(mat_01 , mat_02 ) :
    m , n = mat_01.shape
    p , q = mat_02.shape
    
    if (m!=p) or (n!=q) :
        return 0
    return mat_01*mat_02
    
    
def convolution2D(image , kernel) :
    m , n = image.shape
    p , q = kernel.shape
    out = np.zeros((m-p+1,n-q+1))
    
    for ix in range(out.shape[0] ) :
        for iy in range(out.shape[1]) :
            
            im_patch = image[ix:ix+p , iy:iy +q]
            im_patch = hadamard_prod(im_patch , kernel)
#             s = im_patch.mean()
            s = im_patch.sum()
            if s<0 :
                s = 0.0
            elif s>255.0 :
                s = 255.0
            out[ix,iy] = s
            #convolve here
            
    return out

```

```python
outline_kernel = [[-1,-1,-1] , 
          [-1,8,-1] , 
          [-1,-1,-1]]
identity_kernel = [[-1, -1, -1], 
                   [-1,  8, -1], 
                   [-1, -1, -1]]
blur_kernel = [[0.0625,0.125,0.0625] , 
               [0.125,0.25,0.125] , 
               [0.0625,0.125,0.0625]]
kernel = np.asarray(blur_kernel)


c = convolution2D(x , kernel)
print c.shape
# c = cv2.cvtColor(c,cv2.COLOR_RGB2GRAY)
plt.figure(0)
plt.imshow(x,cmap = 'gray')
plt.axis('off')

plt.figure(1)
plt.imshow(c , cmap = 'gray')
plt.axis ('off')

plt.show()
```
